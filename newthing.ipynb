```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from datetime import datetime
```

```python
df=pd.read_csv('final_dataframe.csv')
df
```

```python

df=df[df['City']=='Adelphi']
df=df[['Year', "Month", 'AVG_Dry_Spell']]
df
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df['AVG_Dry_Spell'], df['AVG_Dry_Spell'], test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(X_train.values.reshape(-1, 1), y_train)
```

```python
X_test.tail(1)
```

```python
next_value = model.predict(X_test.tail(1).values.reshape(-1, 1))
print(f'The predicted value for the next "Avg_Dry_Spell" is {next_value}.')
```

```python
df.loc[len(df.index)] = [2023, 'January', next_value]
df.at[df.index[-1], 'AVG_Dry_Spell'] = df['AVG_Dry_Spell'].iloc[-1][-1]


df
```
