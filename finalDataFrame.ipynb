```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
drySpells = pd.read_csv ('drySpells.csv')
drySpells=drySpells.set_index(['Year', 'Month', 'State', 'City'])
frostDays=pd.read_csv('frostDays.csv')
frostDays=frostDays.set_index(['Year', 'Month', 'State', 'City'])
maxHighTemp=pd.read_csv('maxHighTemp.csv')
maxHighTemp=maxHighTemp.set_index(['Year', 'Month', 'State','City'])
minLowTemp=pd.read_csv('minLowTemp.csv')
minLowTemp=minLowTemp.set_index(['Year', 'Month', 'State','City'])
precipitation=pd.read_csv("precipitation.csv")
precipitation=precipitation.set_index(['Year', 'Month', 'State','City'])
```

```python
df = pd.concat([drySpells, frostDays, maxHighTemp, minLowTemp, precipitation], axis=1)
df
```

```python
adelphi_df = df.reset_index(drop=False)
adelphi_df = adelphi_df.loc[adelphi_df['City']=='Adelphi']
adelphi_df['Date'] = adelphi_df['Year'].astype(str) + '-' + adelphi_df['Month'].astype(str)
adelphi_df = adelphi_df.drop(['Month', 'State', 'City', 'Year'], axis=1).set_index('Date')
adelphi_df.index = pd.to_datetime(adelphi_df.index)

ack_df = df.reset_index(drop=False)
ack_df = ack_df.loc[ack_df['City']=='Ackerman']
ack_df['Date'] = ack_df['Year'].astype(str) + '-' + ack_df['Month'].astype(str)
ack_df = ack_df.drop(['Month', 'State', 'City', 'Year'], axis=1).set_index('Date')
ack_df.index = pd.to_datetime(ack_df.index)

albany_df = df.reset_index(drop=False)
albany_df = albany_df.loc[albany_df['City']=='Albany']
albany_df['Date'] = albany_df['Year'].astype(str) + '-' + albany_df['Month'].astype(str)
albany_df = albany_df.drop(['Month', 'State', 'City', 'Year'], axis=1).set_index('Date')
albany_df.index = pd.to_datetime(albany_df.index)

alachua_df = df.reset_index(drop=False)
alachua_df = alachua_df.loc[alachua_df['City']=='Alachua']
alachua_df['Date'] = alachua_df['Year'].astype(str) + '-' + alachua_df['Month'].astype(str)
alachua_df = alachua_df.drop(['Month', 'State', 'City', 'Year'], axis=1).set_index('Date')
alachua_df.index = pd.to_datetime(alachua_df.index)


# Saving weather data for each city for further analysis
adelphi_df.to_csv('adelphi_df')
ack_df.to_csv('ack_df')
albany_df.to_csv('albany_df')
alachua_df.to_csv('alachua_df')
```
