```python
import requests
import json
import pandas as pd
auth = {"Authorization": "Token e3242fc4e9da547a8adda3abbcf328baa8e2ddd4"}
url = "https://app.climate.azavea.com/api/climate-data/344/RCP85/indicator/max_high_temperature/?time_aggregation=monthly"
resp = requests.get(url, headers=auth)
resp_obj=resp.content
```

```python
data=json.loads(resp_obj)
df=pd.DataFrame(data['data'])
adelphiDrySpells=df.transpose()
adelphiDrySpells.index.name='Year'
adelphiDrySpells=adelphiDrySpells.reset_index()
adelphiDrySpells[['Year', 'Month']] = adelphiDrySpells['Year'].str.split('-', expand=True)
adelphiDrySpells = adelphiDrySpells.reindex(columns=['Year','Month', 'avg'])
adelphiDrySpells[['Year', 'Month']] = adelphiDrySpells[['Year', 'Month']].astype(int)
adelphiDrySpells = adelphiDrySpells[(adelphiDrySpells['Year'] >= 2006) & (adelphiDrySpells['Year'] <= 2023)]
adelphiDrySpells = adelphiDrySpells.iloc[:-6,:]

adelphiDrySpells=adelphiDrySpells.set_index('Year')
month=adelphiDrySpells['Month']
monthList=[]
monthDict = {
    1: 'January',
    2: 'February',
    3: 'March',
    4: 'April',
    5: 'May',
    6: 'June',
    7: 'July',
    8: 'August',
    9: 'September',
    10: 'October',
    11: 'November',
    12: 'December'
}
for num in month:
    month=monthDict[num]
    monthList.append(month)
adelphiDrySpells['Month']=monthList
adelphiDrySpells=adelphiDrySpells.rename(columns={'avg': 'Adelphi_MD'})
adelphiDrySpells=adelphiDrySpells.reset_index()
adelphiDrySpells=adelphiDrySpells.set_index(['Year', 'Month'])
```
